```python
import matplotlib.pyplot as plt
# Data
categories = ['A', 'B', 'C', 'D', 'E']
values = [3, 7, 5, 9, 6]
# Plotting
plt.bar(categories, values, color='blue')
# Labels and Title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Basic Bar Chart')
# Show Plot
plt.show()
```
